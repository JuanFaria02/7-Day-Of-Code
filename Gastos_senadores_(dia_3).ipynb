Para um cientista de dados é importante usar os dados para prever ações e, nesse caso, prever os gastos. Nesse desafio da Alura, a ideia é prever o gasto dos próximos três meses de forma simplificada. A estratégia usada foi a média dos últimos 3 meses para essa previsão. 


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_data_soma  = pd.read_csv('/content/drive/MyDrive/Projetos /Gastos Senadores Brasileiros/Arquivo dados/dataset_ceaps_forecasting.csv')
df_data_soma = pd.DataFrame(df_data_soma)
df_data_soma
```

```python
lista_valores = []


for k in df_data_soma['y']:
  lista_valores.append(k)


lista_valores = lista_valores[::-1]
lista_ultimos_3meses = [i for i in lista_valores]

total  = 0

i = 0
while i != 92:
  total += lista_valores[i]  
  i += 1

media = total / 91

media_meses = media / 3

print(f'Média dos últimos três meses: {media_meses}')
    
```
